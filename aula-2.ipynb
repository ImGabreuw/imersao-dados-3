```python
import pandas as pd

url_dados = "https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true"

dados = pd.read_csv(url_dados, compression = "zip")
dados
```

```python
# Renomear uma coluna do dataframe
mapa = {"droga": "composto"}
dados.rename(columns = mapa, inplace = True)
```

```python
# Pegar os 5 primeiros elementos com mais ocorrência na série
# dados["composto"].value_counts()[0:5] (5 não pertence ao interalo -> Matematicamente [0, 5[)
# ou
# compostos = dados["composto"].value_counts()[:5]
cod_compostos = dados["composto"].value_counts().index[:5]
```

```python
cod_compostos
```

```python
# "@" antes do nome da variável indica que ela foi definida anteriormente (fora da String)
dados.query("composto in @cod_compostos")
```

```python
# Criação de gráficos https://seaborn.pydata.org/
import seaborn as sns
import matplotlib.pyplot as plt

# Salvar as configurações padrão de imagem do Seaborn
sns.set()

# Redimencionar o gráfico
plt.figure(figsize=(8, 6))

ax = sns.countplot(x = "composto", data=dados.query("composto in @cod_compostos"))
ax.set_title("Top 5 compostos")

plt.show()
```
